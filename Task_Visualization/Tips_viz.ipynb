### Step 1. Import the necessary libraries:

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
# set the graphs to show in the jupyter notebook
%matplotlib inline
# set seaborn style to white
sns.set_style("white")
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

```python
address = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv"
tips = pd.read_csv(address)
```

### Step 4. Delete the Unnamed 0 column

```python
tips.columns
```

```python
tips.drop(['Unnamed: 0'], axis='columns', inplace=True)
```

### Step 5. Plot the total_bill column histogram

```python
plt.figure(figsize=(8,4))
sns.distplot(tips['total_bill'], 
            rug = True)
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
sns.jointplot(x = tips['total_bill'],
             y = tips['tip'],
             color = '#4CB391')
```

### Step 7.  Create one image with the relationship of `total_bill`, `tip` and `size`.
#### Hint: It is just one function.

```python
sns.pairplot(tips, kind = 'reg')
```

### Step 8. Present the relationship between days and total_bill value

```python
plt.figure(figsize=(12,8))
sns.swarmplot(x = tips["day"], y = tips["total_bill"], hue = tips['day'], s = 10)
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x = tips["tip"], y = tips["day"], hue = tips['day'], s = 6)
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(12,8))
sns.boxplot(x = tips["time"], y = tips["total_bill"], hue = tips['time'])
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x = tips["day"], y = tips["total_bill"], hue = tips['time'])
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
fig, ax =plt.subplots(1,2)
sns.countplot(x = tips["tip"], hue = tips['time'] == 'Lunch', ax=ax[0])
sns.countplot(x = tips["tip"], hue = tips['time'] == 'Dinner', ax=ax[1])
fig.show()
```

```python
g = sns.FacetGrid(tips, col="time")
g.map_dataframe(sns.histplot, x="tip")
#bien pero no aparece abajo la leyenda de ´tip
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
#muy curiosa la diferencia fumadores vs no fumadores !!!! Lo que se gastan en tabaco no lo dejan en propina jaja
```
