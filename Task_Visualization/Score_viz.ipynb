### Step 2. Create the DataFrame it should look like below.

```python
data = [{'first_name' : 'Jason', 'last_name': 'Miller', 'age' : 42, 'female' : 0, 'preTestScore': 4, 'postTestScore': 25}, {'first_name' : 'Molly', 'last_name': 'Jacobson', 'age' : 52, 'female' : 1, 'preTestScore': 24, 'postTestScore': 94}, {'first_name' : 'Tina', 'last_name': 'Ali', 'age' : 36, 'female' : 1, 'preTestScore': 31, 'postTestScore': 57}, {'first_name' : 'Jake', 'last_name': 'Milner', 'age' : 24, 'female' : 0, 'preTestScore': 2, 'postTestScore': 62}, {'first_name' : 'Amy', 'last_name': 'Cooze', 'age' : 73, 'female' : 1, 'preTestScore': 3, 'postTestScore': 70}]

data = pd.DataFrame(data)
data
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x = data["preTestScore"], y = data["postTestScore"], hue = data['age'], s = 100)
```

```python
sns.relplot(data=data,
            x="preTestScore",
            y="postTestScore",
            hue="age", size="age",
            sizes=(50, 500),
            alpha=.5, palette="muted",
            height=6);
```
