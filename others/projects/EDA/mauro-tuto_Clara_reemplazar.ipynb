```python
import pandas as pd

url = "https://raw.githubusercontent.com/Samanosukeh/Analise-descritiva-de-dados-titanic/master/titanic.csv"
```

```python
sea = pd.read_csv(url)
```

```python
sea.isnull().sum()
```

```python
listaletras = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J", "K", "L"]
```

```python
dicti = {key:value for key, value in zip(sea.columns, listaletras)}
dicti
```

```python
# https://stackoverflow.com/questions/33218719/replace-non-null-values-in-column-by-1
def sustituir(df, diccionario):
    for col in df.columns:
        df.loc[df[col].notnull(), col] = diccionario[col]
    return df  
```

```python
sustituir(sea, dicti)
```
